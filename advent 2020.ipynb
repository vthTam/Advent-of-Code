```python
#Day 3
data = open('data3_2020.txt', 'r').read().split('\n')
data
```

```python
total_trees = 0
height = len(data)
down = 2
right = 1
x_pos = 0
for index in range(down, height-down, down):
    line = data[index]
    x_pos += right
    if x_pos > len(line)-1:
        x_pos = x_pos-len(line)
    if line[x_pos] == "#":
        total_trees+=1
print(total_trees)
```

```python
#Day 4
passports = open('data4_2020.txt', 'r').read().split('\n\n')
valid_passports=0
for p in passports:
    sections = p.split() #splits using space or next line character
    if len(sections) >6:
        byr = False #(Birth Year)
        iyr = False #(Issue Year)
        eyr = False #(Expiration Year)
        hgt = False #(Height)
        hcl = False #(Hair Color)
        ecl = False #(Eye Color)
        pid = False #(Passport ID)
#         display = ""
        for sec in sections:
            code = sec.split(":")[0]
            content = sec.split(":")[1]
            if code == "byr" and int(content) in range(1920,2003):
                byr = True
            elif code == "iyr" and int(content) in range(2010,2021):
                iyr = True
            elif code == "eyr" and int(content) in range(2020,2031):
                eyr = True
            elif code == "hgt" and (content.endswith("cm") or content.endswith("in")):
                height = int(content[:-2])
                if (height in range(150,194) and content.endswith("cm")) or (height in range(59,77) and content.endswith("in")):
                    hgt = True
#                     display=sec
            elif code == "hcl" and content.startswith("#") and len(content)==7:
                hcl = True
            elif code == "ecl" and (content=="amb" or content=="blu" or content=="brn" or content=="gry" or content=="grn" or content=="hzl" or content=="oth"):
                ecl = True
            elif code == "pid" and len(content)==9:
                pid = True
#                 display=sec
        if byr and iyr and eyr and hgt and hcl and ecl and pid:
            valid_passports += 1
#             print(display,"\n")
print("Number of valid passports =",valid_passports)
```

```python
#Day 5
seats = open('data5_2020.txt', 'r').read().split('\n')

# Part 1
highest = -1
all_scores =[] # only used in Part 2
for s in seats:
    rows = range(128) # 0 to 127
    index=0
    while index < 7:
        half = int(len(rows)/2) #length i.e.64,32,16,8,4,2,1
        if s[index]=="F":
            rows = range(rows[0], rows[half])
        elif s[index]=="B":
            rows = range(rows[half], rows[half]+half)
        index += 1
    # only 1 element remains in column
    columns = range(8) # 0 to 7
    while index < 10:
        half = int(len(columns)/2)
        if s[index]=="L":
            columns = range(columns[0], columns[half])
        elif s[index]=="R":
            columns = range(columns[half], columns[half]+half)
        index += 1
    # only 1 element remains in rows
    score = rows[0]*8+columns[0]
    if highest < score:
        highest = score
    all_scores.append(score) # For Part 2's use

print("Highest seat ID:", highest)

# Part 2
all_scores.sort()
for i in range(len(all_scores)):
    if all_scores[i]+1 != all_scores[i+1]:
        print("Missing seat ID:", all_scores[i]+1)
        break
```
