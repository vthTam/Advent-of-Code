```python
#Part 2
data = input
scores = []
for line in data:
  incomplete = True
  opBrackets = []
  for bracket in line:
    if bracket=="(" or bracket=="[" or bracket=="{" or bracket=="<":
      opBrackets.append(bracket)
    elif not opBrackets:
      print("This is not supposed to happen!")
      incomplete = False
      break
    elif bracket==")":
      if opBrackets.pop()!="(":     
        incomplete = False
        break
    elif bracket=="]":
      if opBrackets.pop()!="[":     
        incomplete = False
        break
    elif bracket=="}":
      if opBrackets.pop()!="{":     
        incomplete = False
        break
    elif bracket==">":
      if opBrackets.pop()!="<":     
        incomplete = False
        break
  if incomplete:
    opBrackets.reverse()
    scores.append(calulate(opBrackets))

scores.sort()
print("Middle score:", scores[int(len(scores)/2)])
```

```python
def calulate(brackets):
  total = 0
  for i in brackets:
    if i=="(": 
      total *= 5
      total += 1
    elif i=="[": 
      total *= 5 
      total += 2
    elif i=="{": 
      total *= 5
      total += 3
    elif i=="<": 
      total *= 5
      total += 4
      
  return total  
```
