```python
#Part 2
from collections import deque

basins = 0
data = [[int(x) for x in r] for r in input]
results = [[0] * len(r) for r in data]

for row, nums in enumerate(data):
  for col, num in enumerate(nums):
    # print(data[row][col], num)
    if data[row][col]==9 or results[row][col]!=0:
      continue
    map = {(row,col)}
    queue = deque()
    queue.append((row,col))
    while queue:
      row, col = queue.popleft()
      results[row][col] = basins
      dirs = [(row+1,col), (row-1,col), (row,col-1), (row,col+1)]
      for x, y in dirs:
        if (x, y) in map: 
          continue
        if -1<x<len(data) and -1<y<len(data[x]):
          if data[x][y] != 9:
            map.add((x, y))
            queue.append((x, y))
    basins += 1

sizes = basins*[0]
for line in results:
  for basin in line:
    if basin != 0:
      sizes[basin] += 1

sizes.sort(reverse=True)
print(sizes)
sizes[0]*sizes[1]*sizes[2]
```
